```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
#Get html for name
url = 'https://www.behindthename.com/name/lisa/top/united-states'

#Get results
results=pd.read_html(url)
```

```python
#Write name to df
lisa_name=results[1]
lisa_name
```

```python
#Drop extra columns
lisa_name=lisa_name.drop(columns=[3,4,5,6])

#Drop first 2 rows
lisa_name=lisa_name.drop([0,1])

#Rename columns
lisa_name=lisa_name.rename(columns = {0:'Year', 1:'Rank', 2: 'Percent Used'})

#Replace empty string and - with nan
lisa_name['Rank'].replace('', np.nan, inplace=True)
lisa_name['Rank'].replace('-', np.nan, inplace=True)
lisa_name['Percent Used'].replace('-', np.nan, inplace=True)

#Drop nan
lisa_name=lisa_name.dropna()

#Remove hash from rank
lisa_name['Rank']=lisa_name['Rank'].str[1:]
```

```python
#Write year and rank to integer
lisa_name['Year'] = lisa_name['Year'].astype(int)
lisa_name['Rank'] = lisa_name['Rank'].astype(int)
```
